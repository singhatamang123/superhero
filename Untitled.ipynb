```python
new_customer = new_customer.drop(['Unnamed: 16','Unnamed: 17','Unnamed: 18','Unnamed: 19'], axis = 1)
new_customer
```

```python
new_customer.duplicated().value_counts()
```

```python
# Adding date of birth 

now = pd.Timestamp('now')
now

new_customer['age'] = (now-new_customer['DOB']).astype('<m8[Y]')
new_customer.head()
```

```python
def agegroup(age):
    if age < 30:
        return 'under 30'
    elif age <= 40:
        return '30-39'
    elif age <= 50:
        return '40-49'
    elif age <= 60:
        return '50-59'
    else:
        return '60+'
    
new_customer['age_group'] = new_customer['age'].apply(agegroup)
    
```
